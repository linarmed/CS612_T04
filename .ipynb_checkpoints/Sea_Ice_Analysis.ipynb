```python
import numpy as np
import pandas as pd
import seaborn as sb 
import itertools
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates 
from matplotlib.offsetbox import AnchoredText
import datetime
```

```python
 # read in the data from the provided csv file
df = pd.read_csv('seaice.csv') 
df['Date'] = pd.to_datetime(df[['Year','Month','Day']])
df.index = df['Date'].values
df
```

```python
#Calculate mean for North Hemisphere
NorthH = df[df['Hemisphere'] == 'north']
NorthH_month =  NorthH.resample('1M').mean()
NorthH_month['Month'] = NorthH_month['Month'].apply(lambda x:int(x))
NorthH_month
#Calculate mean for South Hemisphere
SouthH = df[df['Hemisphere'] == 'south']
SouthH_month =  SouthH.resample('1M').mean()
SouthH_month['Month'] = SouthH_month['Month'].apply(lambda x:int(x))
SouthH_month
```

```python
#Heatmap for North hemisphere
plt.figure(figsize=(10,4))
NorthMonthExtent = NorthH_month.pivot("Month", "Year", "Extent")
plt.xticks(rotation=60)
plt.title(' North Hemispher: Mean of Sea Ice Extent')
sb.heatmap(NorthMonthExtent,     cmap='jet_r',annot_kws={"size":4})
 
```

```python
#Heatmap for South hemisphere
plt.figure(figsize=(10,4))
SouthMonthExtent = SouthH_month.pivot("Month", "Year", "Extent")
plt.xticks(rotation=60)
plt.title(' South Hemispher: Mean of Sea Ice Extent')
sb.heatmap(SouthMonthExtent,     cmap='jet_r',annot_kws={"size":4})
```
